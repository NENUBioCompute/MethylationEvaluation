## 数据词典构造

```python
disease_value
```

```python
data_col = ['GEO ID', 'Tissue', 'Disease', 'Condition', 'Age', 'Age_unit', 'Gender', 'Race']
tissue_col = ['GEO ID', 'Unique Value', 'Tissue lvl1', 'Tissue lvl2', 'Tissue lvl3']
disease_col = ['GEO ID', 'Unique Value', 'Disease']
age_col = ['GEO ID', 'Unique Value', 'Age']
gender_col = ['GEO ID', 'Unique Value', 'Gender']
condition_col = ['GEO ID', 'Unique Value', 'Condition']
race_col = ['GEO ID', 'Unique Value', 'Race lvl1', 'Race lvl2']
```

```python
from collections import defaultdict

def get_dict(data, data_col):
    data_df = data[data_col]
    data_dict = defaultdict(list)
    for index in data_df.index:
        if str(data_df.loc[index].values[0]) != 'nan':
            for i, col in enumerate(data_col):
                data_dict[col].append(data_df.loc[index].values[i])
    return data_dict
```

```python
data_dict = get_dict(data, data_col)
tissue_dict = get_dict(tissue_value, tissue_col)
disease_dict = get_dict(disease_value, disease_col)
age_dict = get_dict(age_value, age_col)
gender_dict = get_dict(gender_value, gender_col)
condition_dict = get_dict(condition_value, condition_col)
race_dict = get_dict(race_value, race_col)
```

```python
import json

def write_json_file(obj_path, output):
    with open(obj_path, 'w') as jf:
        json.dump(output, jf, indent=4, ensure_ascii=False)
    return
```

```python
write_json_file('json file/data_dict.json', data_dict)
write_json_file('json file/tissue_dict.json', tissue_dict)
write_json_file('json file/disease_dict.json', disease_dict)
write_json_file('json file/age_dict.json', age_dict)
write_json_file('json file/gender_dict.json', gender_dict)
write_json_file('json file/condition_dict.json', condition_dict)
write_json_file('json file/race_dict.json', race_dict)
```
