```python
import pandas as pd
import json
from collections import defaultdict
```

```python
def read_json(path):
    with open(path, 'r') as jf:
        data = json.load(jf)
    return data
```

```python
data = read_json('json file/data_dict.json')
tissue = read_json('json file/tissue_dict.json')
disease = read_json('json file/disease_dict.json')
condition = read_json('json file/condition_dict.json')
age = read_json('json file/age_dict.json')
gender = read_json('json file/gender_dict.json')
race = read_json('json file/race_dict.json')
```

## 构建融合词表

```python
def filled_dict(mapping_dict, key, obj_dict, item):
    for i, GEO in enumerate(obj_dict['GEO ID']):
        val = list(mapping_dict[GEO][key].keys())[0]
        unique_value = obj_dict['Unique Value'][i] # if str(obj_dict['Unique Value'][i]) != 'nan' else 0
        mapping_dict[GEO][key][val][unique_value] = obj_dict[item][i]
    return mapping_dict
```

```python
mapping_dict = defaultdict(dict)
for i, GEO in enumerate(data['GEO ID']):
    for key in list(data.keys())[1:]:
        mapping_dict[GEO][key] = {data[key][i] :defaultdict(dict)}
    
mapping_dict = filled_dict(mapping_dict, 'Tissue', tissue, 'Tissue lvl3')
mapping_dict = filled_dict(mapping_dict, 'Disease', disease, 'Disease')
mapping_dict = filled_dict(mapping_dict, 'Condition', condition, 'Condition')
mapping_dict = filled_dict(mapping_dict, 'Age', age, 'Age')
mapping_dict = filled_dict(mapping_dict, 'Gender', gender, 'Gender')
mapping_dict = filled_dict(mapping_dict, 'Race', race, 'Race lvl2')
import json

with open('json file/mapping_dict_lv3.json', 'w') as jf:
    json.dump(mapping_dict, jf, indent=4, ensure_ascii=False)
```
